```python
import sys
import os
from matplotlib import pyplot as plt
src=f"{os.getcwd()}/src"
if src not in sys.path:
    sys.path.append(src)
from SimpleEC import *
from operators import *
```

```python
ecb=SimpleEC(eggholder,[(-512,512),(-512,512)], coding='b',
             parents_selection=proportional_selection, population_size=200)
ecg=SimpleEC(eggholder,[(-512,512),(-512,512)], coding='g',population_size=200)
```

```python
avgb,bestb=ecb.evolve()
avgg,bestg=ecg.evolve()
```

```python
fig,ax=plt.subplots(1,1,figsize=(3,3))
ax.plot(bestb, label='Best Binary')
ax.plot(bestg, label='Best Gray')
ax.legend()
```

```python
fig,ax=plt.subplots(1,1,figsize=(3,3))
ax.plot(avgb, label='Avg Binary')
ax.plot(avgg, label='Avg Gray')
ax.legend()
```
